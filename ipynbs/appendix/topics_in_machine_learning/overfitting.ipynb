# 如何处理过拟合问题


首先来明确下机器学习的目标.

机器学习的目标是让机器拥有学习能力,所谓学习其基本可以描述为通过原始信息提取规则.

既然是是学习问题,那么我们可以参考我们自己学习时候的情况做类比:

+ 欠拟合:光看书不做题觉得自己会了，上了考场啥都不会。

+ 过拟合: 做课后题全都能做对，上了考场还是啥都不会。

+ 差不多的拟合: 做了题，背了老师给画了重点，考试60分过了。

+ 优秀的拟合: 课后题全能作对，考试100分。

+ 特征选择就是划重点.

过拟合的内在原因是只训练出了细节而没有总结出规律.就像光看书背课文没有理解其背后的道理一样.而在机器学习这个语境下,通常过拟合是由于:

+ 训练集和测试集特征分布不一致

+ 模型太过复杂而样本量不足.

那么如何处理过拟合问题呢?

## 识别过拟合

首先我们要能够识别模型是否过拟合,这样才能分清要不要使用处理过拟合的方法.要识别过拟合可以通过以下指标:

+ 训练集上准确度远高于验证集上准确度
+ 训练集上准确度还在上升但验证集上准确度已经不再上升或者已经下降了.

通常我们会配合交叉验证来识别过拟合

## 处理过拟合

处理过拟合大体上有这么几种思路:

+ 从数据上提升性能
+ 从算法上提升性能
+ 从模型融合上提升性能

### 从数据上提升性能

1. 做数据标准化
2. 使用数据增强额外的产生新数据用于训练
3. 进行特征工程,比如使用pca是训练时的特征正交


### 从算法上提升性能

1. 通过为一些训练数据设置权重以避免一些特殊点对算法的影响
2. 选取合适的停止训练标准,使对机器的训练在合适的程度.
3. 模型中加入正则项
4. 神经网络中可以增加dropout层随机滤掉一些权重变化
5. 精心挑选权重的初始化的分布

### 从模型融合上提升性能

使用投票,bagging通过组合几个强分类器的结果投票实现提高泛化水平的能力
